```python
import pandas as pd
```

## Load the datasets 

```python
orders = pd.read_csv("../data/olist_orders_dataset.csv")

orders.head()
```

```python
order_items = pd.read_csv("../data/olist_order_items_dataset.csv")

order_items.head()
```

## Method to check dataset information 

```python
def checkDataset(df):
    print("\033[1mShape of the dataset:\033[0m ",df.shape)
    print('-'*100)
    print("\033[1mColumns of the dataset:\033[0m ",orders.columns)
    print('-'*100)
    print("\033[1mDatatype of the columns:\033[0m ", df.dtypes)
    print('-'*100)
    print(df.info())
    print('-'*100)
    print("\033[1mNull values in a column:\033[0m\n",df.isnull().sum())
    print('-'*100)
    print("\033[1mDescription of the dataset:\n\033[0m ",df.describe())
    print('-'*100)
```

```python
checkDataset(orders)
```

## Data Cleaning and Preprocessing for Orders table 

```python
# convert data to datetime64

orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
```

```python
orders.info() # check the datatypes again for orders table
```

```python
print(orders.isnull().sum()) # check for the null values in the column
```

```python
# analyse the order purchase frequency yearly, monthly, daily, hourly

orders['order_year'] = orders['order_purchase_timestamp'].dt.year
orders['order_month'] = orders['order_purchase_timestamp'].dt.month
orders['order_day'] = orders['order_purchase_timestamp'].dt.day
orders['order_hour'] = orders['order_purchase_timestamp'].dt.hour
```

```python
# calculate delivery data

orders['delivery_time']=orders['order_estimated_delivery_date']-orders['order_purchase_timestamp']
```

```python
orders['delivery_time'].describe()
```

```python
orders['delivery_time_days'] = orders['delivery_time'].dt.days # calculate the delivery days in time
```

```python
# remove unnecessary values( if any)

orders[orders['delivery_time_days'] < 0]
```

```python
orders.duplicated().sum()
```

## Data Merging 

```python
orders_items_merged = pd.merge(orders, order_items, on="order_id", how="inner")

orders_items_products = pd.merge(
    orders_items_merged,
    products,
    on="product_id",
    how="left"
)

full_data = pd.merge(
    orders_items_products,
    customers,
    on="customer_id",
    how="left"
)
```

```python
full_data.shape
```

```python
full_data.head()
```

```python
full_data.dtypes
```

```python
full_data['order_month'] = full_data['order_purchase_timestamp'].dt.to_period('M')

full_data['order_year'] = full_data['order_purchase_timestamp'].dt.year
```

## Core Business Analysis 

```python
# total price
full_data['price'].sum()
```

```python
# monthly reveneue trend
monthly_revenue=full_data.groupby('order_month')['price'].sum()

monthly_revenue.plot(figsize=(10,15))
```

```python
# top selling product

top_selling_product= full_data.groupby('product_category_name')['price'].sum().sort_values(ascending=False)
# top_selling_product.plot(figsize=(20,30))
print(top_selling_product)
```

```python
# Orders by state
full_data['customer_state'].value_counts()
```

```python
# average delivery time
full_data['delivery_time_days'].mean()
```

```python
# Top 10 customers
top_customers = full_data.groupby('customer_id')['price'].sum().sort_values(ascending=False)

top_customers.head(10)
```

```python
# orders every month

full_data.groupby('order_month')['order_id'].nunique()
```

```python
# Average order value

avg_order_value= full_data.groupby('order_id')['price'].sum().mean()
print(avg_order_value)
```

```python
# Analyse delivery performance - long long delievries take

full_data['delivery_time_days'].describe()
```

```python
full_data['delivery_time_days'].hist(bins=30)
```

```python
# Product category performance
    
category_orders=full_data.groupby('product_category_name')['order_id'].count().sort_values(ascending=False)
print(category_orders.head(10))
```

```python
# State-wise revenue - how much revenue is generated by each state
    
state_revenue=full_data.groupby('customer_state')['price'].sum().sort_values(ascending=False)
print(state_revenue.head(10))
```

```python
# top sellers

seller_revenue = full_data.groupby('seller_id')['price'].sum().sort_values(ascending=False)
seller_revenue.head(10)
```
